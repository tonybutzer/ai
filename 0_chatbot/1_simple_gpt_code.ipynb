```python
import openai  # Assuming OpenAI provides a GPT-4 API

# Set your OpenAI GPT-4 API key
openai.api_key = 'YOUR_API_KEY'

# Function to generate a response using the GPT-4 API
def generate_response(prompt):
    response = openai.Completion.create(
        model="text-davinci-003",  # Specify the GPT-4 model
        prompt=prompt,
        temperature=0.7,  # Adjust temperature for response creativity
        max_tokens=150,  # Limit the length of the response
        stop=None  # You can specify stop words to control the length further
    )
    return response.choices[0].text.strip()
```

```python
my_prompt='give me a short summary bullets of the jupiter planet'
```

```python
generate_response(my_prompt)
```

```python
#gpt-3.5-turbo
from openai import OpenAI
from dotenv import load_dotenv
import os

# Load environment variables from the .env file
load_dotenv('~/.env')

# Access the OpenAI API key
openai_api_key = os.getenv("OPENAI_API_KEY")

# Now you can use openai_api_key in your code
print("OpenAI API Key:", openai_api_key)

# gets API Key from environment variable OPENAI_API_KEY
client = OpenAI()

# Non-streaming:
print("----- standard request -----")
completion = client.chat.completions.create(
    model="text-davinci-003",
    messages=[
        {
            "role": "user",
            "content": "Say this is a test",
        },
    ],
)
print(completion.choices[0].message.content)
```
